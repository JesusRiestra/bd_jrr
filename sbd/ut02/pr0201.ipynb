```python
import redis

r = redis.Redis(
    host="redis",
    port=6379,
    db=0,
    decode_responses=True
)
```

```python
# 1
r.set("app:version", "1.0")
```

```python
# 2
r.get("app:version")
```

```python
# 3
r.set("app:version", "1.1")
```

```python
# 4
r.set("contador:descargas", 0)
```

```python
# 5
r.incr("contador:descargas", 5)
```

```python
# 6
r.decr("contador:descargas", 2)
```

```python
# 7
r.set("app:estado", "activo")
```
